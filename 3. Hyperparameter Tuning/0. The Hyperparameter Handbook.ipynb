# **THE HYPERPARAMETER HANDBOOK**

### 0.1 Motivation

When training neural networks, various parameters, such as the batch size $B$, learning rate $\eta$, and number of epochs $N$, are chosen before the learning process begins. Setting their exact values is not immediately obvious. These parameters, which are not trainable via backpropagation, are known as **hyperparameters**. Tuning these hyperparameters to their most optimal values is essential to maximising model performance. The process of hyperparameter tuning can often be time-consuming, computationally expensive and technically challenging. However, poor choices of hyperparameters can cause the model to overfit or underfit. This is especially the case with hyperparameters that are directly associated with regularisation techniques that intend to prevent or reduce overfitting. Therefore, hyperparameter tuning is an essential component of the end-to-end DL pipeline, and this notebook aims to be a practical and handy resource for its optimisation.


### 0.2 Challenges of Hyperparameter Optimisation

The challenge of hyperparameter optimisation arises due to the size of the hyperparameter search space. Typically, there are many hyperparameters to be optimised in a deep learning model which leads to a large search space. Moreover, there are many different types of parameters: continuous, discrete or binary which further complicates the search space. Moreover, there is the possibility of multiple local optima in the search space [...]. Paired with lengthy model training times, hyperparameter optimisation becomes an exceptional challenge.

### 0.3 The Linear Scaling Law

As mentioned previously, the batch size, learning rate, and number of training epochs are hyperparameters. In practical settings, these hyperparameters are often the most important. One should not tune each individually as changing one heavily impacts the effect of the other on training. Reviewing the effect of each varying each of these hyperparameters, 

| Hyperparameter     | Typical Values                     | Effect on Training                                                                 |
|-------------------|----------------------------------|-------------------------------------------------------------------------------------|
| Batch Size, $B$        | `8`, `16`, `32`, `64`, `128`, `256`, etc.             | Larger batch sizes lead more stable gradient updates (less stochasticity). When paired with CUDA, it takes advantage of parallel computuation and can lead to faster training. Smaller batches create noisier gradients (more stochasticity), better generalisation, slower training. |
| Learning Rate, $\eta$     | $10^{-5}$ to $10^{-1}$     | Higher learning rate can lead to faster convergence, but risk divergence/overshooting. Lower learning rates are more stable, but have slower convergence and may get stuck in local minima. $\eta$ is heavily dependent on the optimiser, but a common value is $10^{-3}$. |
| Training Epochs $N$   | 10 to 100+   | Increasing the number of epochs allow for more parameter updates which may lead to a better model fit. Too many training epochs lead to overfitting while too few cause underfitting.  |

Perhaps the most important rule when manually tuning these hyperparameters is the **linear scaling rule** proposed by Goyal et. al (2017) [1]. The rule states that: When the (mini)batch size $B$ is multiplied by a integer factor $k$, the learning rate $\eta$ should also be multipled by the same factor such that the new learning rate is $k\eta$. While seemingly simple, this rule allowed them to train `ResNet-50` an order of magnitude quicker by using a batch size of `8192` rather than `256`. For batches sizes under approximately `8,192`, the linear scaling rule holds (provided a gradual warmup period) and is not impacted by the so-called _generalisation gap_ [2]. It is important to note that larger batches result in fewer parameter updates per epoch. However, to maintain the same computational budget and prevent overfitting, the total number of epochs $N$ is typically kept constant, meaning the model sees the same amount of data but performs $1/k$ as many updates. From a high-level, the linear scaling rule arises from the theoretical basis that larger minibatch gradient updates are more reliable since they are computed over more samples. With more reliable gradient updates, gradient-descent based optimisers handle larger update sizes because the stochastic noise is suppressed. This allows the optimiser to take larger "steps" in the direction of the true gradient without overshooting local minima or becoming unstable.
